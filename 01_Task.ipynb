```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
site = "https://www.passiton.com/inspirational-quotes"

req = Request(site)

page = urlopen(req)

soup = BeautifulSoup(page, 'html.parser')

print(soup)
```

```python
themes = soup.find('div', id="all_quotes")

cetagories = themes.select('div h5 a')

for category in cetagories:
    print(category.string)
```

```python
images = themes.select('div img')

for image in images:
    print(image.get('src'))
```

```python
quotes = themes.select('div img')

for quote in quotes:
    print(quote.get('alt').split('#')[0])
```

```python
page = soup.find('div', id="all_quotes")

categories = themes.select('div h5 a')

images = themes.select('div img')

quotes = themes.select('div img')

categories_list = list()
images_list = list()
quotes_list = list()

for category, image, quote in zip(categories, images, quotes):
    categories_list.append(category.string)
    images_list.append(image.get('src'))
    quotes_list.append(image.get('alt').split('#')[0])
```

```python
print(categories_list)
```

```python
print(images_list)
```

```python
print(quotes_list)
```

```python
data = {
    "Category": categories_list,
    "Image": images_list,
    "Quote": quotes_list
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```
